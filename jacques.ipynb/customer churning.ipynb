CUSTOMER CHURNING

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\DELL\Desktop\Customer Churn new.csv")
data.head()
```

CHECKING MISSING DATA

```python
data.info()
```

```python
dataset = data.isna().sum()
print("missing values in each column:\n",dataset)
```

```python
data.shape
data.head()
```

DROPPING COLUMNS CustomerId and Surname

```python
data.columns
```

```python
data.drop(['CustomerId', 'Surname'], axis=1, inplace=True)
data.head()
```

HANDLING CATEGORICAL VARIABLES

GENDER

```python
data['Gender'].unique()
```

```python
data['Gender'] = data['Gender'].apply(lambda x: 0 if x == 'no' else 1)
data.head()
```

GEOGRAPHY

```python
data['Geography'].unique()
```

```python
Geography_dummies = pd.get_dummies(data['Geography'],dtype=int, drop_first=True)
Geography_dummies.head()
```

```python
data = pd.concat([Geography_dummies, data], axis=1)
data.head()
```

```python
data.drop(['Geography'], axis=1, inplace=True)
data.head()
```

STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()   
```

```python
data[['CreditScore', 'Age', 'Tenure', 'Balance', 'Germany', 'Spain', 'RowNumber', 'EstimatedSalary']] = scaler.fit_transform(data[['CreditScore', 'Age', 'Tenure', 'Balance', 'Germany', 'Spain', 'RowNumber', 'EstimatedSalary']]) 
data.head()
```

CREATING A TRAINING SET AND TEST SET

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]    
```

GETTING THE TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)    
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

BUILDING THE MODEL

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

TRAINING MODEL

```python
model.fit(x_train, y_train) 
```

INFERENCE

```python
y_predict = model.predict(x_test)
```

EVALUATING THE MODEL

MAKING CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict) 
```

```python
result = (1546+53) / (1546+53+353+48)
print("Accuracy:", result)   
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
