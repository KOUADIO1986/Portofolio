GETTING THE TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)    
print("x_train shape:", X_train.shape)
print("x_test shape:", X_test.shape)
```

BUILDING AND TRAINING MODEL

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

Reshape X_train to 2D

```python
if len(X_train.shape) == 1:
    X_train = np.expand_dims(X_train, axis=1)  # Reshape to 2D
```

# Convert X_train to a NumPy array, ,, #Convert X_test to a NumPy array, # Ensure y_train is a 1D array and # Ensure y_test to a Numpy array

```python
# Convert X_train to a NumPy array
X_train = X_train.values if isinstance(X_train, pd.DataFrame) else X_train
#Convert X_test to a NumPy array
X_test = X_test.values if isinstance(X_test, pd.DataFrame) else X_test
# Ensure y_train is a 1D array
y_train = y_train.values.ravel() if isinstance(y_train, pd.Series) else y_train
# Ensure y_test to a Numpy array
y_test = y_test.values if isinstance(y_test, pd.DataFrame) else y_test
```

```python
import xgboost
model = xgboost.XGBClassifier(max_depth=4, learning_rate=0.1, n_estimators=100)
model.fit(X_train, y_train)
```

INFERENCE

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

EVALUATING THE MODEL

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```


k-Fold Cross Validation

# Ensure X is a NumPy array and # Ensure y is a 1D array

```python
# Ensure X is a NumPy array
X = X.values if isinstance(X, pd.DataFrame) else X

# Ensure y is a 1D array
y = y.values.ravel() if isinstance(y, pd.Series) else y
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = model,
                             X = X,
                             y = y,
                             scoring = 'accuracy',
                             cv = 10)
print("Average Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
